```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
from scipy.stats import linregress
```

```python
# Study data files
# Read the mouse data and the study results
mouse_data = pd.read_csv("data/Mouse_metadata.csv")
study_data = pd.read_csv("data/Study_results.csv")
```

```python
# Combine the data into a single dataset
mouse_study = pd.merge(mouse_data, study_data, on = ["Mouse ID"])
mouse_study_df = pd.DataFrame(mouse_study)
# Display the data table for preview
mouse_study_df
```

```python
# Checking the number of mice.
len(mouse_study_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = mouse_study_df[mouse_study_df.duplicated(["Mouse ID", "Timepoint"])]
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
true_mouse_study = mouse_study.drop_duplicates(subset = ["Mouse ID", "Timepoint"])
true_mouse_study_df = pd.DataFrame(true_mouse_study)
true_mouse_study_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(true_mouse_study_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
regimen_index = true_mouse_study_df.groupby(["Drug Regimen"])
regimen_index.head()
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
```

```python
# mean of the tumor volume
MEAN = regimen_index["Tumor Volume (mm3)"].mean()
MEAN
```

```python
#median of the tumor volume
MEDIAN = regimen_index["Tumor Volume (mm3)"].median()
MEDIAN
```

```python
#variance of the tumor volume
VAR = regimen_index["Tumor Volume (mm3)"].var()
VAR
```

```python
#standard deviation of the tumor volume
STD = regimen_index["Tumor Volume (mm3)"].std()
STD
```

```python
#SEM of the tumor volume
SEM = regimen_index["Tumor Volume (mm3)"].sem()
SEM
```

```python
# Assemble the resulting series into a single summary dataframe.
stat_summary = ({"MEAN Tumor Volume": MEAN,
                "MEDIAN Tumor Volume": MEDIAN,
                "VARIANCE Tumor Volume": VAR,
                "STD Tumor Volume": STD,
                "SEM Tumor Volume": SEM})
stat_summary_df = pd.DataFrame(stat_summary)
stat_summary_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
Stat_Agg_Summary = regimen_index["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
Stat_Agg_Summary
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_time = true_mouse_study.groupby(["Mouse ID"])["Timepoint"].max()
greatest_time
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_merge = pd.merge(mouse_study_df, greatest_time, on = (["Mouse ID", "Timepoint"]))
timepoint_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

for treatment in treatments:
    treatment_regimen = timepoint_merge.loc[timepoint_merge["Drug Regimen"] == treatment]
    tumor_vol = treatment_regimen['Tumor Volume (mm3)']
    tumor_vol_data.append(tumor_vol)

# Calculate the IQR and quantitatively determine if there are any potential outliers.
quartiles = tumor_vol_data[0].quantile([0.25, 0.5, 0.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq - lowerq
lowerbound = lowerq - (1.5 * iqr)
upperbound = upperq + (1.5 - iqr)

print(f"For {treatments[0]}, values below {lowerbound} and above {upperbound} could be outliers")
```

```python
quartiles1 = tumor_vol_data[1].quantile([0.25, 0.5, 0.75])
lowerq1 = quartiles1[0.25]
upperq1 = quartiles1[0.75]
iqr1 = upperq1 - lowerq1
lowerbound1 = lowerq1 - (1.5 * iqr1)
upperbound1 = upperq1 + (1.5 - iqr1)

print(f"For {treatments[1]}, values below {lowerbound1} and above {upperbound1} could be outliers")
```

```python
quartiles2 = tumor_vol_data[2].quantile([0.25, 0.5, 0.75])
lowerq2 = quartiles2[0.25]
upperq2 = quartiles2[0.75]
iqr2 = upperq2 - lowerq2
lowerbound2 = lowerq2 - (1.5 * iqr2)
upperbound2 = upperq2 + (1.5 - iqr2)

print(f"For {treatments[2]}, values below {lowerbound2} and above {upperbound2} could be outliers")
```

```python
quartiles3 = tumor_vol_data[3].quantile([0.25, 0.5, 0.75])
lowerq3 = quartiles3[0.25]
upperq3 = quartiles3[0.75]
iqr3 = upperq3 - lowerq3
lowerbound3 = lowerq3 - (1.5 * iqr3)
upperbound3 = upperq3 + (1.5 - iqr3)

print(f"For {treatments[3]}, values below {lowerbound3} and above {upperbound3} could be outliers")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Tumor Growth results by Treatment')
ax1.set_ylabel('Growth (mm3)')
ax1.set_xlabel("Treatment")
ax1.boxplot(tumor_vol_data, labels = treatments)
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
Capo_scatter = true_mouse_study.loc[true_mouse_study["Drug Regimen"] == "Capomulin"]
Capo_scatter
```

```python
tumor_scatter = Capo_scatter.groupby(["Mouse ID"]).mean()
tumor_scatter.head()
```

```python
avg_tumor = tumor_scatter["Tumor Volume (mm3)"]
avg_tumor
```

```python
avg_weight = tumor_scatter["Weight (g)"]
avg_weight
```

```python
plt.scatter(avg_weight, avg_tumor, marker="o", facecolor = "blue", edgecolors = "black")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight of mouse on Capomulin vs Tumor volume")
plt.show()
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(avg_weight, avg_tumor)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")
```

```python
x_values = tumor_scatter["Weight (g)"]
y_values = tumor_scatter["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept, 2))
plt.annotate(line_eq,(20,36), fontsize = 15, color = 'red')
plt.plot(x_values, regress_values, "r-")



plt.scatter(avg_weight, avg_tumor, marker="o", facecolor = "blue", edgecolors = "black")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight of mouse on Capomulin vs Tumor volume")

plt.show()
```
