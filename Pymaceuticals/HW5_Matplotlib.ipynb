```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
```

```python
# Study data files
# Read the mouse data and the study results
mouse_data = pd.read_csv("data/Mouse_metadata.csv")
study_data = pd.read_csv("data/Study_results.csv")
```

```python
# Combine the data into a single dataset
mouse_study = pd.merge(mouse_data, study_data, on = ["Mouse ID"])
mouse_study_df = pd.DataFrame(mouse_study)
# Display the data table for preview
mouse_study_df
```

```python
# Checking the number of mice.
len(mouse_study_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = mouse_study_df[mouse_study_df.duplicated(["Mouse ID", "Timepoint"])]
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
true_mouse_study = mouse_study.drop_duplicates(subset = ["Mouse ID", "Timepoint"])
true_mouse_study_df = pd.DataFrame(true_mouse_study)
true_mouse_study_df
```

```python
# Checking the number of mice in the clean DataFrame.
len(true_mouse_study_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
regimen_index = true_mouse_study_df.groupby(["Drug Regimen"])
regimen_index.head()
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
```

```python
# mean of the tumor volume
MEAN = regimen_index["Tumor Volume (mm3)"].mean()
MEAN
```

```python
#median of the tumor volume
MEDIAN = regimen_index["Tumor Volume (mm3)"].median()
MEDIAN
```

```python
#variance of the tumor volume
VAR = regimen_index["Tumor Volume (mm3)"].var()
VAR
```

```python
#standard deviation of the tumor volume
STD = regimen_index["Tumor Volume (mm3)"].std()
STD
STD.dtypes
```

```python
#SEM of the tumor volume
SEM = regimen_index["Tumor Volume (mm3)"].sem()
SEM
```

```python
# Assemble the resulting series into a single summary dataframe.
stat_summary = ({"MEAN Tumor Volume": MEAN,
                "MEDIAN Tumor Volume": MEDIAN,
                "VARIANCE Tumor Volume": VAR,
                "STD Tumor Volume": STD,
                "SEM Tumor Volume": SEM})
stat_summary_df = pd.DataFrame(stat_summary)
stat_summary_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
Stat_Agg_Summary = regimen_index["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
Stat_Agg_Summary
```

# Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
Timepoints = regimen_index["Timepoint"].count()

bar_time = Timepoints.plot.bar(x = "Drug Regimen", y = "Timepoint")
bar_time.set_title("Timepoints for each Drug Regimen")
bar_time.set_xlabel("Drug Regimen")
bar_time.set_ylabel("Timepoints")
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
x_axis = np.arange(len(Timepoints))
y_axis = Timepoints

tick_location = [value for value in x_axis]
plt.xticks(tick_location, ["Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol", \
                         "Placebo", "Propriva", "Ramicane", "Stelasyn", "Zoniferol"], rotation = "75")

plt.bar(x_axis, y_axis, color = "red", alpha = 0.5, align = "center")
plt.title("Timepoints for each Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Timepoints")
plt.show()
```
